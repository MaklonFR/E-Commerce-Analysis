* **Mengubah Tipe Data**

```python
datetime_columns = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']

for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
```

Pada tahap ini, karena missing value pada kolom order_approved_at, order_delivered_carrier_date dan order_estimated_delivery_date tidak terlalu banyak maka kita akan menghapus missing value dengan menggunakan method `fillna()`.

```python
# order_approved_at, order_delivered_carrier_date dan order_estimated_delivery_date
# Tentukan kolom yang ingin dicek (misalnya tiga kolom tertentu)
columns_to_check = ['order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date']

# Hapus baris yang memiliki missing value pada kolom tertentu
orders_df = orders_df.dropna(subset=columns_to_check)
```

```python
orders_df.isna().sum()
```

Hasil diatas menunjukan tabel orders_df sudah tidak memiliki nilai null.

## Exploratory Data Analysis (EDA)

### Eksplorasi Data oders_df

Pada langkah ini, kita akan mengeksplorasi data oders_df terlebih dahulu. Kita akan menggunakan method apply() untuk melakukan sebuah operasi terhadap setiap elemen dalam sebuah kolom DataFrame atau Series (bentuk satu dimensi dari DataFrame). Operasi yang akan kita lakukan ialah menghitung jumlah detik dari **order_delivered_customer_date** menggunakan method total_seconds(). Nilai tersebut selanjutnya diubah ke dalam satuan hari (dibagi 86400) dan diambil bilangan bulatnya saja.

```python
order_delivered_customer_time = orders_df["order_delivered_customer_date"] - orders_df["order_delivered_carrier_date"]
order_delivered_customer_time = order_delivered_customer_time.apply(lambda x: x.total_seconds())
orders_df["order_delivered_customer_time"] = round(order_delivered_customer_time/3600)
orders_df.sort_values(by='order_delivered_customer_time', ascending=False).head(5)
```

```python
order_delivered_customer_late = orders_df["order_estimated_delivery_date"] - orders_df["order_delivered_customer_date"]
order_delivered_customer_late = order_delivered_customer_late.apply(lambda x: x.total_seconds())
orders_df["order_delivered_customer_late"] = round(order_delivered_customer_late/86400)
orders_df.sort_values(by='order_delivered_customer_time', ascending=False).head(5)
```

## Analisis Lanjutan (Opsional)

### Analisis RFM

**Analisis RFM** adalah teknik pemasaran yang mengelompokkan pelanggan berdasarkan kekinian, frekuensi, dan nilai moneter transaksi mereka.

- Keterkinian: Mengukur seberapa baru pelanggan melakukan pembelian.
- Frekuensi: Menganalisis seberapa sering pelanggan melakukan pembelian.
- Nilai Moneter: Mengevaluasi total nilai moneter dari transaksi pelanggan.

#### Gabungan RFM (Recency, Frequency dan Monetary)

```python
rfm_df = merged_df.groupby(by="customer_unique_id", as_index=False).agg({
    "order_purchase_timestamp": "max", # mengambil tanggal order terakhir
    "order_id": "nunique", # menghitung jumlah order
    "payment_value": "sum" # menghitung jumlah revenue yang dihasilkan
})
rfm_df.columns = ["customer_unique_id", "max_order_timestamp", "frequency", "monetary"]

# menghitung kapan terakhir pelanggan melakukan transaksi (hari)
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"]
recent_date = orders_df["order_purchase_timestamp"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
rfm_df.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

colors = ["#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4", "#72BCD4"]

sns.barplot(y="recency", x="customer_unique_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette='dark:#5A9_r', legend=False, hue='customer_unique_id', ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis ='x', labelsize=15)

sns.barplot(y="frequency", x="customer_unique_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette='dark:#5A9_r', legend=False, hue='customer_unique_id', ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)

sns.barplot(y="monetary", x="customer_unique_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette='dark:#5A9_r', legend=False, hue='customer_unique_id', ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)

plt.suptitle("Best Customer Based on RFM Parameters (customer_id)", fontsize=20)
plt.show()
```
